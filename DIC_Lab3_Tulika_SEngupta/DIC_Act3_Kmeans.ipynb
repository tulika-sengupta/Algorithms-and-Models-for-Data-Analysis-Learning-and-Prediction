<html><b>Reading data.</b></html>

```r
data <- read.csv(file= "June 10-July 12, 2015 - Gaming, Jobs and Broadband - CSV.csv")
```

<html><b>Selecting five columns.</b></html>

```r
data <- data[,c("age","sex","emplnw","par","inc")]

#"date1a", "emptype1","emptype3",
```

```r
data <- na.omit(data)
```

<html><b>Categorising age.</b></html>

```r
data$ageCat <- cut(data$age,c(18,25,50,80))
head(data)
```

```r
na.omit(data)
```

<html><b>converting data to numeric.</b></html>

```r
data$ageCat <- as.numeric(data$ageCat)
```

<html><b>Labeling the age category.</b></html>

```r
if((data$ageCat > 18 ) && (data$ageCat <=25)){
    data$ageCat <- 1
}
if((data$ageCat > 25 ) && (data$ageCat <=50)){
    data$ageCat <- 2
}
if((data$ageCat > 50 ) && (data$ageCat <=80)){
    data$ageCat <- 3
}
```

```r
data$sex <- as.numeric(data$sex)
```

```r
data <-data[,c("age","sex","emplnw","par","inc")] 
data <- na.omit(data)
```

<html><b>Normalizing Data.</b></html>

```r
data$age <- (data$age - min(data$age)) /(max(data$age)-min(data$age))
data$sex <- (data$sex - min(data$sex)) /(max(data$sex)-min(data$sex))
data$emplnw <- (data$emplnw - min(data$emplnw)) /(max(data$emplnw)-min(data$emplnw))
data$par <- (data$par - min(data$par)) /(max(data$par)-min(data$par))
data$inc <- (data$inc - min(data$inc)) /(max(data$inc)-min(data$inc))
```

```r
data
```

<html><b>Performing k-means for number of clusters 3.</b></html>

```r
k <-kmeans(data, centers=3, algorithm=c("Hartigan-Wong")) #Create 3 clusters
```

```r
rng<-2:20 #K from 2 to 20
tries<-100 #Run the K Means algorithm 100 times
avg.totw.ss<-integer(length(rng)) #Set up an empty vector to hold all of points
```

<html><b>Taking the two clustering values.</b></html>

```r
v <- c(3,5)
```

<html><b>Trying other values for K</b></html>

```r
for(j in 1:2){ # For each value of the range variable
v.totw.ss<-integer(tries) #Set up an empty vector to hold the 100 tries
for(i in 1:tries){
k.temp<-kmeans(data,centers=v[j]) #Run kmeans
v.totw.ss[i]<-k.temp$tot.withinss#Store the total withinss
}
avg.totw.ss[v-1]<-mean(v.totw.ss) #Average the 100 total withinss
}
```

<html><b>Performing k-means for number of clusters 5.</b></html>

```r
k <-kmeans(data, centers=5,algorithm=c("Hartigan-Wong"))
table(k$cluster)
k$centers
```
