<html><b>Reading German Data</b></html>

```r
data <- read.csv("german_credit_data1.csv")
#data <- data[,c("Age","Sex","Job","Housing","Credit.amount","Duration","Purpose","Credit.Risks")]
class(data)
str(data)
#data
```

```r
data <- rename(data, c("Credit Risks"="Credit.Risks"))
data
```

<html><b>Replacing the NA values and converting the data to numeric type and normalizing the data.</b></html>

```r
data$Saving.accounts <- factor(data$Saving.accounts, exclude = NULL, 
               levels = c("little", "moderate", "quite rich", "rich", NA), 
               labels = c(1, 2, 3, 4, 0))

data$Checking.account <- factor(data$Checking.account, exclude = NULL, 
               levels = c("little", "moderate", "rich", NA), 
               labels = c(1, 2, 3, 0))

data$Sex <- factor(data$Sex, exclude = NULL, 
               levels = c("male", "female"), 
               labels = c(1, 2))

data$Housing <- factor(data$Housing, exclude = NULL, 
               levels = c("free", "own","rent"), 
               labels = c(1, 2,3))

data$Purpose <- factor(data$Purpose, exclude = NULL, 
               levels = c("business", "car", "domestic appliances", "education", "furniture/equipment", "radio/TV", "repairs", "vacation/others"), 
               labels = c(1, 2,3,4,5,6,7,8))

#converrting data to numeric type
data$Sex <- as.numeric(data$Sex)
data$Saving.accounts <- as.numeric(data$Saving.accounts)
data$Checking.account <- as.numeric(data$Checking.account)
data$Housing <- as.numeric(data$Housing)
data$Purpose <- as.numeric(data$Purpose)


#normalizing data
data$Sex <- (data$Sex - min(data$Sex)) /(max(data$Sex)-min(data$Sex)) 
data$Purpose <- (data$Purpose - min(data$Purpose)) /(max(data$Purpose)-min(data$Purpose)) 
data$Saving.accounts <- (data$Saving.accounts - min(data$Saving.accounts)) /(max(data$Saving.accounts)-min(data$Saving.accounts)) 
data$Checking.account <- (data$Checking.account - min(data$Checking.account)) /(max(data$Checking.account)-min(data$Checking.account)) 
data$Housing <- (data$Housing - min(data$Housing)) /(max(data$Housing)-min(data$Housing)) 


data
```

```r
n.points <- nrow(data)# number of rows in the dataset
```

<html><b>Taking three values for sampling rate.</b></html>

```r
sampling.rate <- c(0.8,0.7,0.6)
```

<html><b>Obtaining the test labels .</b></html>

```r
#num.test.set.labels <- n.points * (1 - sampling.rate)
for (i in 1:3) 
   { num.test.set.labels[i] <- n.points * (1 - sampling.rate[i])}
```

<html><b>Performing KNN and calculating misclassification rate.</b></html>

```r
for(i in 1:3)
   {
    cat("\n Sample rate:", sampling.rate[i])
    training <- sample(1:n.points, sampling.rate[i] * n.points,replace=FALSE)
    train <- subset(data[training, ], select=c("Age","Sex","Job","Housing","Saving.accounts","Checking.account","Credit.amount","Duration","Purpose"))
   
    # the other rows are going into the test set
    testing <- setdiff(1:n.points, training)
    # define the test set to be the other rows
    test <- subset(data[testing, ], select=c("Age","Sex","Job","Housing","Saving.accounts","Checking.account","Credit.amount","Duration","Purpose"))
    #print(test)
    
    cl <- data$Credit.Risks[training]
    # this is the subset of labels for the training set
    true.labels <- data$Credit.Risks[testing]
    # subset of labels for the test set, we're withholding these
   # cl
    cat("\n K                                Miscalssification Rate")
    for(k in 1:20) {
      predicted.labels <- knn(train,test,cl,k)
      num.incorrect.labels<-sum(predicted.labels !=true.labels)
      misclassification.rate<-num.incorrect.labels/num.test.set.labels[i]
      cat("\n",k,"\t \t \t \t \t",misclassification.rate)
}
    
    }
```
